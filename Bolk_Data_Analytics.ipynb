```python
from google.colab import auth
import pandas as pd

auth.authenticate_user()

query = """
with tmp as (
  select distinct Order_ID, Recipe, Payment_date from `bolk-data-consulting-371411.bolk.orders` where Recipe is not null
)


SELECT Customer_ID as customer_id 
,o.Order_ID as order_id
,Order_group_ID as order_group_id
,Order_success as order_success
,Order_group_state as order_group_state
,Robot_ID as robot_id
,Robot as robot
,t.Payment_date as payment_date
,t.Recipe as recipe
,Super_Bowl as super_bowl
,Ingredient as ingredient
,Ingredient_ID as ingredient_id
,Weight as weight
,Distributed as distributed
,Price as price
,ROUND(Total_price,2) as total_price
,ROUND(Refunded,2) as refunded
,Discount_tag as discount_tag
,ROUND(Used_cagnotte,2) as used_cagnotte
,if (Green_hours_discount_amount is not null,0.5,0) as green_hours_discount_amount
,Is_menu_discounted as is_menu_discounted
,Nayax_transaction_id as nayax_transaction_id
,Nayax_card_number as nayax_card_number
,Nayax_card_bin as nayax_card_bin
FROM `bolk-data-consulting-371411.bolk.orders` o
LEFT JOIN tmp t on o.order_id=t.Order_ID 
"""

df = pd.read_gbq(query,project_id="bolk-data-consulting-371411")
df
```
